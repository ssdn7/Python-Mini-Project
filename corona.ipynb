```python
import numpy as np
import pandas as pd
import datetime 
import matplotlib.pyplot as plt
from datetime import date, datetime, time, timezone

df = pd.read_csv('./data/covid-vaccination-doses-per-capita.csv')
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Day'])
df.set_index('Date', inplace=True)
df.drop(['Day'], axis=1, inplace=True)
df.head()
```

```python
len(df['Entity'].unique())
```

```python
covid_c = df.groupby(['Entity'])
```

```python
for key, group in covid_c:
    print('+key: ', key)
    print('+number: ', len(group))
    print(group.head()) 
    print('\n')
```

```python
total_df = covid_c.sum()
total_df.head()
```

```python
import folium

map = folium.Map(location=[37.2594750011864, 127.012324585], zoom_start=13)
map
```

```python
import json
center = [35.949123124555, 84.01293758433]

m = folium.Map(location=center, zoom_start=2, max_bounds=True, min_zoom=1, min_lat=-84, max_lat=84, min_lon = -175, max_lon=187)
geo_path = './data/world-countries.json'
json_data = json.load(open(geo_path), encoding='utf-8')

folium.Choropleth(geo_data=json_data,
                  data=total_df,
                  columns=(total_df.index, 'total_vaccinations_per_hundred'),
                  fill_color=0.7,
                  line_opacity=0.5,
                  ).add_to(m)
folium.LayerControl().add_to(m)
```
